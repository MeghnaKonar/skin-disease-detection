```python
import os,cv2
from PIL import Image
import matplotlib.image as mpimg
import numpy as np
import cv2
from pylab import imshow, show
```

```python
path = '/content/drive/MyDrive/DataSet/train/Warts Molluscum and other Viral Infections/' # Source Folder
dstpath = '/content/drive/MyDrive/black_white_images/warts/' # Destination Folder
```

```python
try:
    os.mkdir(dstpath)
except:
    print ("Directory already exist, images will be written in asme folder")
```

```python
for image in os.listdir(path):
    img = cv2.imread(os.path.join(path,image))
    gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    cv2.imwrite(os.path.join(dstpath,image),gray)
```
